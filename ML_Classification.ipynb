```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
## Data for clustering
x1_a = np.random.normal(1, 0.5, 1000)
x2_a = np.random.normal(1, 0.5, 1000)
temp_x = np.random.uniform(-1, 1, 150)
temp_y = np.random.uniform(-1, 1, 150)
temp_r = temp_x**2 + temp_y**2
temp_x = temp_x[temp_r < 1]
temp_y = temp_y[temp_r < 1]
temp_r = temp_r[temp_r < 1]
temp_x /= temp_r**0.5
temp_y /= temp_r**0.5
x1_n = temp_x + np.random.normal(1, 0.1, temp_x.shape)
x2_n = temp_y + np.random.normal(1, 0.1, temp_y.shape)
```

```python
plt.plot(x1_a, x2_a, '.')
plt.plot(x1_n, x2_n, '.')
```

```python
plt.plot(x1_a, x2_a, '.')
```

```python
plt.plot(x1_n, x2_n, '.')
```

```python
x1 = np.concatenate([x1_a, x1_n])
x2 = np.concatenate([x2_a, x2_n])
y = np.concatenate([np.ones(x1_a.shape), np.zeros(x1_n.shape)])
data = np.stack([y, x1, x2])
data = data.T
np.random.shuffle(data)
```
