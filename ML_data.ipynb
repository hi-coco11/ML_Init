```python
n = 100000
x = np.random.uniform(-1, 1, n)
y = np.random.uniform(-1, 1, n)

r = np.sqrt(x**2 + y**2)
x = x/r
y = y/r

plt.xlim([-2, 2])
plt.ylim([-2,2])
plt.plot(x, y, '.')
```
